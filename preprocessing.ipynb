```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pickle
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df.head(6)
```

```python
df.drop(columns=["RowNumber", "CustomerId","Surname"],inplace=True)
```

```python
df.head()
```

```python
labelenc = LabelEncoder()

df["Gender"] = labelenc.fit_transform(df["Gender"])
```

```python
df.head()
```

```python
oneHotenc = OneHotEncoder()
encoded_geo = oneHotenc.fit_transform(df[["Geography"]])
encoded_geo = encoded_geo.toarray()
```

```python
oneHotenc.get_feature_names_out()
```

```python
df2 = pd.DataFrame(encoded_geo, columns=oneHotenc.get_feature_names_out())
```

```python
df2
```

```python
df.drop(["Geography"],axis=1,inplace=True)
df = pd.concat([df,df2],axis=1)
```

```python
df.head()
```

```python
with open("labelenc.pkl","wb") as file:
    pickle.dump(labelenc,file)

with open("oneHotenc.pkl","wb") as file:
    pickle.dump(oneHotenc,file)
```

```python
df.to_csv("prep_churn_modelling.csv",index=False)
```
